```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
  sum = 0
  for i in numbers:
    if i % 2 == 0:
      sum += i
  return sum

numbers = [1, 2, 3, 4, 5, 6]
evenSum = sum_of_even_numbers(numbers)
print("The sum of Even Numbers is: ", evenSum)
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(s):
  return s[::-1]

string = "Hello World"
reversed = reverse_string(string)
print(f'The reverse of the string is: {reversed}')
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
  squares = []
  for i in numbers:
    squares.append(i**2)
  return squares

numbers = [1, 2, 3, 4, 5]
squares = square_numbers(numbers)
print(f'The squares of the numbers are: {squares}')
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

number = list(range(1, 201))

def is_prime(number):
  if number <= 1:
    return False
  for i in range(2, number):
    if number % i == 0:
      return False
  return True

print("prime numbers between 1 t0 201 are: ")
for num in number:
  if is_prime(num):
    print(num, end = " ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci_sequence(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a + b

for i in fibonacci_sequence(10):
  print(i, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range(n):
    yield i**2

for i in power_of_two(10):
  print(i, end=" ")
```

```python
# 7.  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
  with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes leading/trailing whitespaces

file_path = 'your_file.txt'  # Replace with your actual file path
for i in read_file_line_by_line(file_path):
    print(i)
```

3. **What are the different ways to define and call a function in Python?**

-> 1. Standard Function Definition

We can define a function using the def keyword followed by the function name and parentheses. Inside the parentheses, you can specify parameters.

def my_function(param1, param2):
    return param1 + param2

result = my_function(5, 3) # calling the function
print(result)  # Output: 8

> 2. Function with Default Parameters
We can define default values for parameters. If the caller does not provide a value for that parameter, the default value is used.

def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# Calling the function
print(greet("Alice"))          # Output: Hello, Alice!
print(greet("Bob", "Hi"))      # Output: Hi, Bob!

> 3.  Variable-Length Arguments
We can use *args to accept a variable number of positional arguments and **kwargs for keyword arguments.

def add_numbers(*args):
    return sum(args)

# Calling the function
print(add_numbers(1, 2, 3, 4))  # Output: 10

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Calling the function
print_info(name="Alice", age=30)  # Output: name: Alice, age: 30

> 4. Lambda Functions
Lambda functions are small anonymous functions defined using the lambda keyword. They can take any number of arguments but can only have one expression.

square = lambda x: x * x

# Calling the lambda function
print(square(4))  # Output: 16

> 5. Nested Functions
You can define a function inside another function. The inner function can access variables from the outer function.

def outer_function(msg):
    def inner_function():
        print(msg)
    inner_function()

# Calling the outer function
outer_function("Hello from the inner function!")  # Output: Hello from the inner function!

5.  **What are iterators in Python and how do they differ from iterables?**

--> An iterator is an object that allows traversal through a countable number of values. It is used to iterate over iterable objects like lists, tuples, sets, etc.

An iterable is any Python object that can return its elements one at a time. This means that we can loop over it using a `for` loop or use it in other contexts that require iteration.

Differences:

> Iterable: An object that can be iterated over (e.g., lists, tuples, strings).
Iterator: An object that keeps track of its current position during iteration and produces the next value when requested.

> Iterable: Implements the `__iter__()` method, which returns an iterator.      
Iterator: Implements both `__iter__()` and `__next__()` methods.

> Iterable: Does not maintain any state about the iteration.          
Iterator: Maintains state (the current position) and can be exhausted (i.e., once it has gone through all its elements, it cannot be reused).

> in Iteratable we can use an iterable in a for loop directly.
in Iterator we need to call `next()` on an iterator to get the next value.

# An iterable
my_list = [1, 2, 3]

# Getting an iterator from the iterable
my_iterator = iter(my_list)

# Using the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3


6.  **Explain the concept of generators in Python and how they are defined.**

They are defined using functions but instead of using the `return` statement to return a value, they use the `yield` statement to yield a value. This allows the function to maintain its state between calls, making it possible to produce a series of values over time rather than computing them all at once and returning them in a list.

7.  **What are the advantages of using generators over regular functions?**

--> The advantages of using generators over regular functions include:

>Memory efficiency: Generators use less memory by yielding one item at a time.

def large_range(n):
    for i in range(n):
        yield i

# Using the generator
for number in large_range(10**6):
    pass  # Only one number is in memory at a time

>Lazy evaluation: Values are computed only when needed, improving performance.

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Only the first few values are generated
for number in count_up_to(5):
    print(number)  # Outputs: 1, 2, 3, 4, 5

>Simplified code: Generators are easier to write and maintain than custom iterator classes.

# Generator function
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Using the generator
for num in fibonacci(10):
    print(num)

>State retention: Generators maintain their state between calls, allowing for complex iteration.

def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Using the generator
for number in countdown(5):
    print(number)  # Outputs: 5, 4, 3, 2, 1

>Infinite sequences: Generators can represent infinite sequences without consuming memory.

def infinite_numbers():
    n = 0
    while True:
        yield n
        n += 1

# Using the generator
for number in infinite_numbers():
    if number > 10:
        break
    print(number)  # Outputs: 0, 1, 2, ..., 10

>Pipelining: Generators can be composed to create efficient data processing pipelines.

def square(numbers):
    for n in numbers:
        yield n * n

def filter_even(numbers):
    for n in numbers:
        if n % 2 == 0:
            yield n

# Using the pipeline
numbers = range(10)
squared_evens = filter_even(square(numbers))

for num in squared_evens:
    print(num)  # Outputs: 0, 4, 16, 36, 64

9. **Explain the purpose and usage of the `map()` function in Python.**

--> The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.

**Purpose of `map()`**

Transformation: `map()` is used to apply a transformation function to each element of an iterable, producing a new iterable with the transformed values.

Conciseness: It provides a more concise way to apply a function to a sequence compared to using a loop.

Functional Programming: It aligns with functional programming paradigms, allowing for cleaner and more readable code when performing operations on collections of data.

**Usage**

>1. Using `map()` with a Single Iterable

numbers = [1, 2, 3, 4, 5]

# Define a function to square a number
def square(x):
    return x ** 2

# Use map to apply the square function to each element in the list
squared_numbers = map(square, numbers)

# Convert the map object to a list to see the results
squared_list = list(squared_numbers)
print(squared_list)  # Output: [1, 4, 9, 16, 25]

>2. Using map() with a Lambda Function

numbers = [1, 2, 3, 4, 5]

# Use map with a lambda function to square the numbers
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert to a list
squared_list = list(squared_numbers)
print(squared_list)  # Output: [1, 4, 9, 16, 25]

>3. Using map() with Multiple Iterables

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Use map to add corresponding elements from both lists
summed = map(lambda x, y: x + y, list1, list2)

# Convert to a list
summed_list = list(summed)
print(summed_list)  # Output: [5, 7, 9]



10.  **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

-->  `map()`:
> Purpose: Applies a function to each item in an iterable (like a list, tuple, etc.) and returns a new iterable with the results.

> Input: Takes a function and one or more iterables as arguments.

> Output: Returns a map object containing the transformed elements. You can convert it to a list using `list()`

numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x**2, numbers))  # Output: [1, 4, 9, 16, 25].

`filter()`:

>Purpose: Filters elements from an iterable based on a given function that returns True or False.

>Input: Takes a function and an iterable as arguments.

>Output: Returns a filter object containing only the elements for which the function returned True. We can convert it to a list using `list()`.

numbers = [1, 2, 3, 4, 5]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # Output: [2, 4]


`reduce()`:

>Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

>Input: Takes a function and an iterable as arguments. You might need to import it from functools: from functools import reduce.

>Output: Returns a single value, which is the result of the cumulative operation.

from functools import reduce

numbers = [1, 2, 3, 4, 5]

product = reduce(lambda x, y: x * y, numbers)  # Output: 120

> map() transforms elements.

> filter() selects elements.

> reduce() combines elements.
